```python
import pandas as pd

filename = list(uploaded.keys())[0]
print(filename)

df = pd.read_csv(filename, encoding="latin1")
df.head()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

df["order_date"] = pd.to_datetime(df["order_date"])
df["ship_date"] = pd.to_datetime(df["ship_date"])

df["total_sales"] = df["sales"] * df["quantity"]

df = df.dropna()

df.to_csv("cleaned_superstore.csv", index=False)

print("Cleaned data saved.")
```

```python
import sqlite3

conn = sqlite3.connect("superstore.db")

df.to_sql("sales", conn, if_exists="replace", index=False)

print("Loaded into SQL database.")
```

```python
query = """
SELECT region, ROUND(SUM(total_sales),2) AS total_sales
FROM sales
GROUP BY region
ORDER BY total_sales DESC
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT product_name, ROUND(SUM(total_sales),2 ) AS revenue
FROM sales
GROUP BY product_name
ORDER BY revenue DESC
LIMIT 10
"""

pd.read_sql(query, conn)
```

```python
query = """
SELECT segment, ROUND(SUM(total_sales),2) AS profit
FROM sales
GROUP BY segment
ORDER BY total_sales DESC
"""

pd.read_sql(query, conn)
```
